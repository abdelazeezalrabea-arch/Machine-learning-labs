```python
# Import Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
```

```python
# Load Dataset

df = pd.read_csv("student-mat.csv")

# Display first 5 rows
df.head()
```

```python
# Check all values
print(df.isna())
```

```python
print(df.isna().sum())
```

```python
# checking duplicate rows
df.duplicated()[df.duplicated()==True]
```

```python
# finding number of rows and columns

print("Shape (rows, columns): ", df.shape,"\n")

print("number of rows: ", df.shape[0])
print("number of columns: ", df.shape[1])
```

```python

df.dtypes
```

```python
#Statistical summary
df.describe(include='all') 
```

```python
plt.figure(figsize=(8,5))
# Using 'G3' (Final Grade) instead of 'Boxes Shipped'
sns.histplot(df['G3'], bins=20) 
plt.title("Distribution of Final Grades (G3)")
plt.show()
```

```python
plt.figure(figsize=(8,5))

# Use 'G3' (Final Grade) instead of 'Amount'
sns.histplot(df['G3'], bins=20, kde=True) 

# Update the title to reflect the data
plt.title("Distribution of Final Grades (G3)")
plt.xlabel("Final Grade (0-20)")
plt.ylabel("Number of Students")

plt.show()
```

```python
job_grades = df.groupby('Mjob')['G3'].mean().sort_values(ascending=False)

plt.figure(figsize=(10,5))
job_grades.plot(kind='bar')
plt.title("Average Final Grade by Mother's Job")
plt.ylabel("Average Final Grade (G3)")
plt.xticks(rotation=0) # Keeps the labels horizontal
plt.show()

job_grades
```

```python
reason_absences = df.groupby('reason')['absences'].sum().sort_values(ascending=False)

plt.figure(figsize=(10,5))
reason_absences.plot(kind='bar')
plt.title("Total Absences by School Choice Reason")
plt.ylabel("Total Absences")
plt.xticks(rotation=0) 
plt.show()

print(reason_absences)
```

```python
age_absences = df.groupby('age')['absences'].sum().sort_values(ascending=False)

print(age_absences.head(10))
```

```python
plt.figure(figsize=(8,5))


sns.scatterplot(x='G1', y='G3', data=df)

plt.title("First Period Grade (G1) vs Final Grade (G3)")
plt.xlabel("First Period Grade (G1)")
plt.ylabel("Final Grade (G3)")

plt.show()
```

```python
cols_to_use = ['studytime', 'failures', 'absences', 'G1', 'G3']

corr_matrix = df[cols_to_use].corr()

plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix of Student Performance")
plt.show()
```

```python
age_trend = df.groupby('age')['G3'].mean()

# 2. Plot the trend
plt.figure(figsize=(10,5))
age_trend.plot()  # Default plot type for Series is a line plot
plt.title("Trend of Final Grades by Age")
plt.ylabel("Average Final Grade (G3)")
plt.xlabel("Age")
plt.grid(True) # Added grid for better readability
plt.show()
```
